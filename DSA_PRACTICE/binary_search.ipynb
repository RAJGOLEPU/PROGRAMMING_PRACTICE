# **BINARY SEARCH PRACTICE & IMPLIMENTATION**

```python
l = [i for i in range(1,100000001)] # list with 100000000 items
```

```python
t = 100000000
```

```python
def normal_linear_search(l,t):
  for i in l:
    if i == t:
      return l.index(i)
```

```python
%%time
normal_linear_search(l,t)
```

```python
def binary_search(my_list, target):
    lo, hi = 0, len(my_list) - 1

    while lo <= hi:
        mid = (lo + hi) // 2

        if my_list[mid] == target:
            # Continue searching in the left half to find the first occurrence
            hi = mid - 1
        elif my_list[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1

    # Check if the target was found
    if lo < len(my_list) and my_list[lo] == target:
        return lo
    else:
        return -1

```

```python
%%time
binary_search(l,t)
```
